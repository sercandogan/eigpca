```python
from eigpca import PCA
from sklearn.datasets import load_iris
```

## Data
We will use Iris data for example.

```python
X = load_iris().data
```

```python
X
```

```python
pca = PCA()
```

```python
pca.fit(X) # could be used fit_transform() for applying fiting and transformation
```

## Explained Variance of Components

```python
pca.explained_variance_ratio_
```

## Eigen values of data

```python
pca.eig_vals
```

```python
pca.plot(y="pov")
```

```python
pca.plot(y="eig")
```

```python
pca.transform(X, n_components=2)
```

```python
pca.transform(X) # automatic number of components selection 
```
